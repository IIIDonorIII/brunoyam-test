```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

url = 'http://mfd.ru/currency/?currency=USD'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table', {'class': 'mfd-table mfd-currency-table'})
rows = table.find_all('tr')

dates = []
rates = []

for row in rows[1:]:
    cells = row.find_all('td')
    date = cells[0].text.strip()
    rate = float(cells[1].text.strip().replace(',', '.'))
    dates.append(date)
    rates.append(rate)
    print(f'{date}: {rate}')

# разворачиваем списки с датами и курсами методом reverse()
dates = list(reversed(dates))
rates = list(reversed(rates))

plt.plot(dates, rates)
plt.xlabel('Дата')
plt.ylabel('Курс доллара (Руб)')
plt.title('Изменение курса доллара в рублях за последние 30 дней')

plt.show()
```

```python
#Модуль 7 уровень 2
import matplotlib.pyplot as plt
import numpy as np

# Создание массива значений X на интервале от -5 до 5 с шагом 0.1
x = np.arange(-5, 5, 0.1)

# Вычисление значений Y для каждого значения X
y = x**2

# Создание графика
plt.plot(x, y)

# Добавление заголовка и подписей осей
plt.title("График функции y = x^2")
plt.xlabel("Значения X")
plt.ylabel("Значения Y")

# Отображение графика
plt.show()
```
